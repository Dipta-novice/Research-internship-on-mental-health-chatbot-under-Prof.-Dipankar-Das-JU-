# Preprocessing for LSTM

```python
# Preprocessing

import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
```

```python
nltk.download('stopwords')
```

```python
ps = PorterStemmer()

corpus = []
for i in x['questionText']:
    i = re.sub('[^a-zA-Z]',' ',i)
    i = i.lower().split()
    i = [ps.stem(j) for j in i if not j in stopwords.words('english')]
    i = ' '.join(i)
    corpus.append(i)
```

```python
import tensorflow
```

```python
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
vocabulary_size = 10000
onehot = [one_hot(i, vocabulary_size) for i in corpus]
```

```python
onehot_pad = pad_sequences(onehot, padding='pre', maxlen=100)
onehot_pad
```

```python
onehot_pad[0]
```

```python
onehot_pad.shape
```

```python
from imblearn.over_sampling import SMOTE
```

```python
sm = SMOTE(k_neighbors=2)
```

```python
x, y = sm.fit_resample(onehot_pad,y)
```

```python
# Creating Model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SpatialDropout1D, LSTM, Dense
from tensorflow.keras.optimizers import Adam
```

```python
neurons_in_last_layer = y.topic.nunique()
```

```python
neurons_in_last_layer
```

```python
# train test split

from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(np.array(x), np.array(y), test_size= 0.3, random_state= 42)
```

```python
import tensorflow as tf
tf.config.experimental_run_functions_eagerly(True)
```

```python
from keras.utils.np_utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
y_test
```

```python
y_test.shape
```

```python
y_t = np.argmax(y_test, axis=1)
y_t
```

#LSTM 3

```python
features = 512

model3 = Sequential()
model3.add(Embedding(vocabulary_size, features, input_length=100))
model3.add(SpatialDropout1D(0.4))
model3.add(LSTM(512, dropout=0.4 ,return_sequences=True))
model3.add(LSTM(256, dropout=0.3))
model3.add(Dense(31, activation='softmax'))

model3.compile(optimizer = Adam(0.0008), loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
model3.summary()
```

```python
model3.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=32)
```

```python
y_pred3 = model3.predict(x_test)
```

```python
y_p3 = np.argmax(y_pred3, axis=1)
y_p3
```

```python
import seaborn as sbn
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(15,10))
df_cm3 = pd.DataFrame(confusion_matrix(y_t, y_p3))
sbn.heatmap(df_cm3, annot=True)
plt.show()
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_t, y_p3)
```

# LSTM using Keras-Tuner

```python
features = 512
def build(hp):
    lstm = Sequential()
    lstm.add(Embedding(vocabulary_size, features, input_length= 100))
    
    for i in range(hp.Int('hidden_layers', min_value=1, max_value=3)):
      lstm.add(LSTM(units= hp.Int('neurons'+str(i), min_value=16, max_value=512, step=32), return_sequences=True))

    lstm.add(LSTM(units= hp.Int('neurons_', min_value=16, max_value=256, step=32)))

    lstm.add(Dense(units=neurons_in_last_layer, activation='softmax'))


              
    lstm.compile(optimizer = Adam(learning_rate = hp.Choice('learning_rate', values=[0.0006, 0.0008, 0.001])), loss = 'categorical_crossentropy', metrics = ['accuracy'])
    return lstm
```

```python
pip install keras_tuner
```

```python
import keras_tuner
from kerastuner import RandomSearch

tuner_search = RandomSearch(
    build,
    objective='val_accuracy',
    max_trials=10,
    executions_per_trial=1,
    directory='output',
    project_name='chatbot'
)
```

```python
tuner_search.search(x_train, y_train, epochs=4, validation_data=(x_test, y_test), batch_size=32)
```

```python
best_lstm = tuner_search.get_best_models(num_models=1)[0]
```

```python
best_lstm.summary()
```

```python
best_lstm.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=10, batch_size=32)
```

```python
y_keras_pred = best_lstm.predict(x_test)
```

```python
y_ = np.argmax(y_keras_pred, axis=1)
y_
```

```python
import seaborn as sbn
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(15,10))
df_cm_ = pd.DataFrame(confusion_matrix(y_t, y_))
sbn.heatmap(df_cm_, annot=True)
plt.show()
```
